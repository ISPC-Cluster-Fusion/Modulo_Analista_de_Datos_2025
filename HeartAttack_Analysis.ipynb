###  **Descripción del Proyecto.**

El presente análisis exploratorio tiene como objetivo principal identificar patrones clínicos asociados a la presencia o ausencia de **infarto agudo de miocardio (IAM)**, a partir del estudio estadístico de un conjunto de datos reales obtenidos en el **hospital Zheen** (Erbil, Iraq), durante el período **enero–mayo de 2019**.

El conjunto de datos incluye variables biomédicas relevantes, como:

* **Edad**
* **Género**
* **Frecuencia cardíaca**
* **Presión arterial sistólica y diastólica**
* **Nivel de glucosa en sangre**
* **Valores de CK-MB y troponina** (marcadores comúnmente utilizados en la práctica clínica para la evaluación del daño miocárdico)

###  **Objetivos del Análisis.**

A través de un enfoque **cuantitativo**, se pretende:

1. **Explorar tendencias y correlaciones** entre variables clínicas y la ocurrencia de IAM.

3. **Evaluar la calidad y completitud de los datos**, con el fin de proponer mejoras en la recolección de datos clínicos.
4. **Establecer las bases para el desarrollo futuro de un modelo predictivo**, que pueda ser utilizado como herramienta de apoyo a la toma de decisiones médicas.

```python
# Análisis columnas numéricas.
# Columnas numéricas a analizar
columnas_numericas = ['Age', 'Heart rate', 'Systolic blood pressure', 'Diastolic blood pressure','Blood sugar', 'CK-MB', 'Troponin']

analisis_numerico = []

for columna in columnas_numericas:
    media = df[columna].mean()
    mediana = df[columna].median()
    std = df[columna].std()
    nulos = df[columna].isnull().sum()
    total = len(df[columna])

    # Valores no numéricos (inválidos)
    registros_invalidos = df[~df[columna].apply(lambda x: isinstance(x, (int, float, np.integer, np.float64)))][columna].unique().tolist()

    analisis_numerico.append({
        "Columna": columna,
        "Valores válidos": total - nulos,
        "Valores nulos": nulos,
        "Media": round(media, 2),
        "Mediana": round(mediana, 2),
        "Desviación estándar": round(std, 2),
        "Valores inválidos": registros_invalidos
    })

# Convertir a DataFrame y mostrar con estilo
analisis_df = pd.DataFrame(analisis_numerico)

analisis_df.style.background_gradient(cmap='Greens').format({
    "Media": "{:.2f}",
    "Mediana": "{:.2f}",
    "Desviación estándar": "{:.2f}",
    "Valores válidos": "{:.0f}",
    "Valores nulos": "{:.0f}"
})
```

```python
# Análisis de columnas categóricas
columnas_categoricas = ['Gender', 'Result']
analisis_categorico = []

for columna in columnas_categoricas:
    conteo = df[columna].value_counts(dropna=False).to_dict()
    valores_nulos = df[columna].isnull().sum()
    valores_invalidos = None
    if columna == 'Gender':
        valores_invalidos = df[~df[columna].isin([0, 1])][columna].unique().tolist()
    if columna == 'Result':
        valores_invalidos = df[~df[columna].isin(['positive', 'negative'])][columna].unique().tolist()
    analisis_categorico.append({
        'Columna': columna,
        'Conteo por categoría': conteo,
        'Valores nulos': valores_nulos,
        'Valores fuera de lo esperado': valores_invalidos
    })

# Convertir en DataFrame
analisis_df_categorico = pd.DataFrame(analisis_categorico)

# Estilizar y mostrar Resultados
analisis_df_categorico.style.set_properties(**{'text-align': 'left'}) \
    .background_gradient(cmap='Purples') \
    .format({'Valores nulos': '{:.0f}'})
```

### 1.1 · Datos ausentes – Ejemplo 1
Situación encontrada La columna age presentaba 1 valor faltante 'NaN', lo que podía distorsionar las estadísticas de edad.

Criterio elegido Reemplazamos el dato vacío por 58 años, que es la mediana (el valor del medio al ordenar las edades). Esto evita que edades muy bajas o muy altas distorsionen el resultado y, además, no tenemos que borrar la fila.

```python
print(df.columns.tolist())
```

### 1.2 · Datos ausentes – Ejemplo 2

Situación encontrada:
Se detectaron valores ausentes en la columna Age (edad). Por ejemplo, en la fila 72 no se registra la edad del paciente, lo cual es relevante para los análisis de riesgo.

Criterio elegido:
Se utilizó la mediana para reemplazar el valor faltante, ya que es una medida robusta frente a valores extremos, especialmente útil en distribuciones no simétricas como suele ser la edad.

```python
# Calc mediana de la edad
mediana_edad = df['Age'].median()

# Guardar los registros con edad nula antes de la imputación
filas_con_Age_nulo = df[df['Age'].isna()].copy()  # usamos .copy() para conservar la vista previa
print("Registros con edad ausente antes de la imputación:")
display(filas_con_Age_nulo)

# reemplazamos valores ausentes cn la mediana
df['Age'] = df['Age'].fillna(mediana_edad)

# verificamos cuantos valores null quedan
print("Valores nulos restantes en 'Age':", df['Age'].isna().sum())

# print de los valores corregidos
filas_corregidas = df.loc[filas_con_Age_nulo.index]
print("\n Registros con edad ausente corregidos con la mediana:")
display(filas_corregidas)
```

Datos ausentes - Ejemplo 3:
Valores faltantes en resultados

```python
# Función para inferir resultado
def inferir_resultado(fila):
    if pd.isnull(fila["Result"]):
        # Reglas de ejemplo (pueden ajustarse con base clínica más precisa)
        if (
            fila["Heart rate"] > 100 or
            fila["Systolic blood pressure"] > 140 or
            fila["Blood sugar"] > 200 or
            fila["CK-MB"] > 5 or
            fila["Troponin"] > 0.01
        ):
            return "positive"
        else:
            return "negative"
    else:
        return fila["Result"]

# Aplicar la función fila por fila
df["Result"] = df.apply(inferir_resultado, axis=1)
```

### 1.3 · Formato incorrecto – Ejemplo 1


Situación encontrada:
Se detectaron valores atípicos o inverosímiles en la columna Age, específicamente registros con edades negativas o superiores a 120 años, lo cual se considera un error de formato o ingreso de datos.

Criterio elegido:
Los valores "?" se consideran erróneos y serán tratados como datos faltantes (NaN) para luego imputarse con la mediana, ya que la mediana es menos sensible a valores extremos que el promedio.

```python
# Reemplazar edades mayores a 120 con una edad promedio razonable (80 años)
df.loc[df['Age'] > 120, 'Age'] = 80

# Confirmar reemplazo
print(" Registros corregidos con valor 80:")
display(df.loc[df['Age'] == 80].reset_index(drop=True))
```

### 1.4 · Formato incorrecto – Ejemplo 2 (`Valores no numéricos en 'Heart rate`)

Situación encontrada:
Se detectaron registros en la columna "Heart rate" que contenían valores no numéricos, los cuales impiden realizar análisis cuantitativos.

Criterio elegido:
Se convirtieron los valores a tipo numérico, utilizando pd.to_numeric con errors='coerce', lo cual transforma los valores no válidos en NaN. Posteriormente, los valores nulos se reemplazaron con la mediana de los valores válidos, garantizando así la integridad del campo sin eliminar datos.

```python
# Buscar valores no numéricos en 'Heart rate'
print(" Valores con formato incorrecto en 'Heart rate':")
formato_incorrecto_hr = df[~df['Heart rate'].apply(lambda x: str(x).isdigit())]
display(formato_incorrecto_hr)

# Reemplazar con la mediana válida
df['Heart rate'] = pd.to_numeric(df['Heart rate'], errors='coerce')
mediana_hr = df['Heart rate'].median()
df['Heart rate'] = df['Heart rate'].fillna(mediana_hr)

print(f" Valores corregidos en 'Heart rate' con la mediana: {mediana_hr}")
```

Formato incorrecto - Ejemplo 3:

Situación encontrada:
Se detectaron valores abreviados e incompletos en la columna Result, tales como "neg", "pos" y valores ausentes (vacíos). Esta situación puede afectar el análisis, ya que impide agrupar correctamente los casos en categorías consistentes.

Criterio elegido:
Se decidió reemplazar las abreviaciones "neg" por "negative" y "pos" por "positive" para estandarizar los datos. Posteriormente, se abordarán los valores ausentes mediante inferencia basada en otras variables clínicas del dataset (frecuencia cardíaca, presión arterial, azúcar en sangre, etc.).

```python
# Reemplazar valores abreviados
df["Result"] = df["Result"].replace({
    "neg": "negative",
    "pos": "positive"
})
```

### 1.5 · Dato erroneo – Ejemplo 1

Dato erróneo: valores negativos en "Diastolic blood pressure"
Situación encontrada:
Se detectaron valores negativos en la presión diastólica, lo cual no es fisiológicamente posible en condiciones humanas normales.

Criterio elegido:
Se reemplazaron estos valores con la mediana de los valores válidos (≥ 0), ya que representa una estimación robusta del centro de la distribución, evitando el sesgo de outliers o errores extremos.

```python
# Presión diastólica no puede ser negativa
print("Valores negativos en 'Diastolic blood pressure':")
errores_diastolica = df[df['Diastolic blood pressure'] < 0]
display(errores_diastolica)


# Reemplazar por la mediana válida
mediana_diastolica = df[df['Diastolic blood pressure'] >= 0]['Diastolic blood pressure'].median()
# Convertir la columna a float para evitar conflictos con la mediana decimal
df['Diastolic blood pressure'] = df['Diastolic blood pressure'].astype(float)


# Reemplazar valores negativos por la mediana
df.loc[df['Diastolic blood pressure'] < 0, 'Diastolic blood pressure'] = mediana_diastolica
print(f"Reemplazados con la mediana válida: {mediana_diastolica}")
```

### 1.6 · Dato erroneo – Ejemplo 2

Situación encontrada:
Se identificó un valor erróneo en la columna "Blood sugar" donde aparecen registros por encima de 250. En condiciones clínicas típicas, un valor de glucosa en sangre por encima de 250 mg/dL puede ser considerado una hiperglucemia grave y en muchos casos extremo o improbable en contextos normales.

Criterio elegido:
Se reemplazan estos valores extremos por la mediana de los valores válidos (≤250 mg/dL), ya que es una medida robusta ante outliers.

```python
# Buscar valores extremos de glucosa en sangre (>250)
datos_erroneos_glucosa = df[df['Blood sugar'] > 250]

print("Valores erróneos detectados en 'Blood sugar' (mayores a 250 mg/dL):")
display(datos_erroneos_glucosa[['Blood sugar']].head(10))  # mostrar los primeros 10 para no saturar

# Reemplazar con la mediana de valores válidos (≤250)
mediana_glucosa_valida = df[df['Blood sugar'] <= 250]['Blood sugar'].median()
df.loc[df['Blood sugar'] > 250, 'Blood sugar'] = mediana_glucosa_valida

# Mostrar confirmación
print(f"\n✅ Valores mayores a 250 reemplazados con la mediana válida: {mediana_glucosa_valida}")
```

### 1.7 · Datos duplicados – Ejemplo 1: Registros completamente duplicados

Situación encontrada:
Se identificaron registros completamente duplicados, es decir, filas con valores idénticos en todas las columnas.

Criterio elegido:
Se eliminaron estos duplicados utilizando drop_duplicates(), lo cual conserva solo una instancia de cada registro duplicado, asegurando integridad sin redundancia.


```python
# Verificar duplicados completos
duplicados_completos = df[df.duplicated()]
print(f"Total de duplicados completos: {duplicados_completos.shape[0]}")
df = df.drop_duplicates()
print("Duplicados completos eliminados.")
```

### 1.8 · Datos duplicados – Ejemplo 2: Duplicados por columnas clave

Situación encontrada:
Se detectaron registros duplicados en el subconjunto de columnas clave (Age, Gender, Heart rate, Blood sugar, CK-MB), es decir, filas que compartían exactamente esos mismos valores en dichos campos.

Criterio elegido:
Se eliminaron los duplicados dejando solo la primera fila con esa combinación de valores, para que cada paciente aparezca una sola vez en esa sección del análisis.

```python
# Duplicados por columnas clave
duplicados_parciales = df[df.duplicated(subset=['Age', 'Gender', 'Heart rate', 'Blood sugar', 'CK-MB'], keep=False)]
print("Duplicados por columnas clave (Age, Gender, Heart rate, 'Blood sugar', 'CK-MB):")
display(duplicados_parciales)

# Eliminar duplicados dejando el primero
df = df.drop_duplicates(subset=['Age', 'Gender', 'Heart rate', 'Blood sugar', 'CK-MB'])
print("Duplicados parciales eliminados.")
```

La mayoría de los pacientes se concentra en el grupo con diagnóstico positivo, aunque se observa una distribución diferenciada por género. Los hombres presentan una mayor proporción de casos positivos en comparación con las mujeres.

#3- Realizar las medidas descriptivas de las variables en base a su tipo. Realizar los informes según sean los datos (simétricos o asimétricos).

##Medidas descriptivas
Se calculan estadísticas para variables numéricas y proporciones para categóricas.

```python
num_cols = [
    'Age','Heart rate','Systolic blood pressure','Diastolic blood pressure','Blood sugar','CK-MB','Troponin'
            ]

desc = df[num_cols].describe().T
display(desc)

cat_cols = ['Gender', 'Result']

for col in cat_cols:
    print(f'\nDistribución de {col}')
    display(
        df[col]
          .value_counts(normalize=True)
          .rename('proporción')
          .to_frame()
          .style.format('{:.1%}')
    )
```

Situación encontrada Queríamos saber, de un vistazo, como se comportan las variables numericas (edad, presión, glucemia, enzimas) y cuán equilibradas están las variables categóricas.

Criterio elegido

Usamos describe() para obtener los números básicos: cantidad de datos, promedio, dispersión, mínimos, máximos y cuartiles.

Para las categóricas, calculamos la proporción de cada categoría con value_counts(normalize=True).

Hallazgos

La edad se concentra alrededor de 58 años.

La presión sistólica media es 127 mm Hg.

El 66 % de los registros corresponden a género codificado como “1” (hombres).

El 61 % de los pacientes tienen resultado positivo en el test de ataque cardíaco.

```python
# MEDIDAS DESCRIPTIVAS BÁSICAS
# Columnas numéricas
num_cols = df.select_dtypes(include="number").columns

for col in num_cols:
    print(f"─── {col} (numérica) ───")
    print("Media            :", df[col].mean())
    print("Mediana          :", df[col].median())
    print("Moda             :", df[col].mode()[0])
    print("Desv. estándar   :", df[col].std())

    # Análisis simetría
    if df[col].mean() > df[col].median():
        print("→ Asimetría positiva (cola a la derecha)")
    elif df[col].mean() < df[col].median():
        print("→ Asimetría negativa (cola a la izquierda)")
    else:
        print("→ Distribución simétrica")
    print()

# Columnas categóricas
cat_cols = df.select_dtypes(exclude="number").columns

for col in cat_cols:
    print(f"─── {col} (categórica) ───")
    print("Categorías únicas :", df[col].nunique())
    print("Moda (más repetido):", df[col].mode()[0])
    print()
```

Situación encontrada Queríamos resumir rápidamente qué tan grandes, variados y balanceados son los datos numéricos y categóricos del dataset.

Criterio elegido

Variables numéricas

Calculamos media, mediana, moda y desvío estándar para ver tendencia central y dispersión.

Comparamos media vs. mediana:

Si son parecidas → la distribución es simétrica.

Si la media es mayor → cola hacia la derecha (asimetría positiva).

Si la media es menor → cola hacia la izquierda (asimetría negativa).

Variables categóricas

Contamos cuántas categorías únicas hay (nunique) y cuál se repite más (mode).

Esto muestra qué clase domina y si el reparto está equilibrado.

Lectura simple

Una desviación estándar alta indica valores muy dispersos; baja, que están más agrupados.

Con las proporciones de gender y result vemos, por ejemplo, que ~66 % de los registros son hombres y ~61 % tienen resultado positivo.

Saber simetría y dispersión nos ayuda a elegir la medida correcta (media o mediana) y a detectar posibles valores atípicos antes de avanzar.
