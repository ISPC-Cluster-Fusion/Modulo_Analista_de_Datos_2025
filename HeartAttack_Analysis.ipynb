```python
# An谩lisis columnas num茅ricas.
# Columnas num茅ricas a analizar
columnas_numericas = ['Age', 'Heart rate', 'Systolic blood pressure', 'Diastolic blood pressure','Blood sugar', 'CK-MB', 'Troponin']

analisis_numerico = []

for columna in columnas_numericas:
    media = df[columna].mean()
    mediana = df[columna].median()
    std = df[columna].std()
    nulos = df[columna].isnull().sum()
    total = len(df[columna])

    # Valores no num茅ricos (inv谩lidos)
    registros_invalidos = df[~df[columna].apply(lambda x: isinstance(x, (int, float, np.integer, np.float64)))][columna].unique().tolist()

    analisis_numerico.append({
        "Columna": columna,
        "Valores v谩lidos": total - nulos,
        "Valores nulos": nulos,
        "Media": round(media, 2),
        "Mediana": round(mediana, 2),
        "Desviaci贸n est谩ndar": round(std, 2),
        "Valores inv谩lidos": registros_invalidos
    })

# Convertir a DataFrame y mostrar con estilo
analisis_df = pd.DataFrame(analisis_numerico)

analisis_df.style.background_gradient(cmap='Greens').format({
    "Media": "{:.2f}",
    "Mediana": "{:.2f}",
    "Desviaci贸n est谩ndar": "{:.2f}",
    "Valores v谩lidos": "{:.0f}",
    "Valores nulos": "{:.0f}"
})
```

```python
# An谩lisis de columnas categ贸ricas
columnas_categoricas = ['Gender', 'Result']
analisis_categorico = []

for columna in columnas_categoricas:
    conteo = df[columna].value_counts(dropna=False).to_dict()
    valores_nulos = df[columna].isnull().sum()
    valores_invalidos = None
    if columna == 'Gender':
        valores_invalidos = df[~df[columna].isin([0, 1])][columna].unique().tolist()
    if columna == 'Result':
        valores_invalidos = df[~df[columna].isin(['positive', 'negative'])][columna].unique().tolist()
    analisis_categorico.append({
        'Columna': columna,
        'Conteo por categor铆a': conteo,
        'Valores nulos': valores_nulos,
        'Valores fuera de lo esperado': valores_invalidos
    })

# Convertir en DataFrame
analisis_df_categorico = pd.DataFrame(analisis_categorico)

# Estilizar y mostrar Resultados
analisis_df_categorico.style.set_properties(**{'text-align': 'left'}) \
    .background_gradient(cmap='Purples') \
    .format({'Valores nulos': '{:.0f}'})
```

### 1.2 路 Datos ausentes  Ejemplo 2

Situaci贸n encontrada:
Se detectaron valores ausentes en la columna Age (edad). Por ejemplo, en la fila 72 no se registra la edad del paciente, lo cual es relevante para los an谩lisis de riesgo.

Criterio elegido:
Se utiliz贸 la mediana para reemplazar el valor faltante, ya que es una medida robusta frente a valores extremos, especialmente 煤til en distribuciones no sim茅tricas como suele ser la edad.

```python
# Calc mediana de la edad
mediana_edad = df['Age'].median()

# Guardar los registros con edad nula antes de la imputaci贸n
filas_con_Age_nulo = df[df['Age'].isna()].copy()  # usamos .copy() para conservar la vista previa
print("Registros con edad ausente antes de la imputaci贸n:")
display(filas_con_Age_nulo)

# reemplazamos valores ausentes cn la mediana
df['Age'] = df['Age'].fillna(mediana_edad)

# verificamos cuantos valores null quedan
print("Valores nulos restantes en 'Age':", df['Age'].isna().sum())

# print de los valores corregidos
filas_corregidas = df.loc[filas_con_Age_nulo.index]
print("\n Registros con edad ausente corregidos con la mediana:")
display(filas_corregidas)
```

Datos ausentes - Ejemplo 3:
Valores faltantes en resultados

```python
# Funci贸n para inferir resultado
def inferir_resultado(fila):
    if pd.isnull(fila["Result"]):
        # Reglas de ejemplo (pueden ajustarse con base cl铆nica m谩s precisa)
        if (
            fila["Heart rate"] > 100 or
            fila["Systolic blood pressure"] > 140 or
            fila["Blood sugar"] > 200 or
            fila["CK-MB"] > 5 or
            fila["Troponin"] > 0.01
        ):
            return "positive"
        else:
            return "negative"
    else:
        return fila["Result"]

# Aplicar la funci贸n fila por fila
df["Result"] = df.apply(inferir_resultado, axis=1)
```

### 1.5 路 Dato erroneo  Ejemplo 1

Dato err贸neo: valores negativos en "Diastolic blood pressure"
Situaci贸n encontrada:
Se detectaron valores negativos en la presi贸n diast贸lica, lo cual no es fisiol贸gicamente posible en condiciones humanas normales.

Criterio elegido:
Se reemplazaron estos valores con la mediana de los valores v谩lidos ( 0), ya que representa una estimaci贸n robusta del centro de la distribuci贸n, evitando el sesgo de outliers o errores extremos.

```python
# Presi贸n diast贸lica no puede ser negativa
print("Valores negativos en 'Diastolic blood pressure':")
errores_diastolica = df[df['Diastolic blood pressure'] < 0]
display(errores_diastolica)


# Reemplazar por la mediana v谩lida
mediana_diastolica = df[df['Diastolic blood pressure'] >= 0]['Diastolic blood pressure'].median()
# Convertir la columna a float para evitar conflictos con la mediana decimal
df['Diastolic blood pressure'] = df['Diastolic blood pressure'].astype(float)


# Reemplazar valores negativos por la mediana
df.loc[df['Diastolic blood pressure'] < 0, 'Diastolic blood pressure'] = mediana_diastolica
print(f"Reemplazados con la mediana v谩lida: {mediana_diastolica}")
```

## Distribuci贸n de diagn贸stico por g茅nero.

```python
# Distribuci贸n de diagn铆stico por g茅nero.

# Eliminar filas donde Gender es NaN
df = df.dropna(subset=['Gender'])

# Verificar que ya no hay nulos
print(df['Gender'].isnull().sum())


# Colores definidos antes
colores = {
    'positive_Mujer': '#F1948A',    # Rojo claro
    'positive_Hombre': '#C0392B',   # Rojo oscuro
    'negative_Mujer': '#3498DB',    # Azul claro
    'negative_Hombre': '#21618C'    # Azul oscuro
}

# Conteo total por resultado para torta
conteo_result = df['Result'].value_counts()

# Conteo por grupo para barras
# Crear columna 'Grupo' combinando Result y Gender
df["Grupo"] = df.apply(lambda fila: f"{fila['Result']}_{'Mujer' if fila['Gender'] == 0 else 'Hombre'}", axis=1)
# Conteo por grupo para barras
conteo_grupo = df['Grupo'].value_counts().reindex(colores.keys())

fig, axs = plt.subplots(1, 2, figsize=(14,6))

# Gr谩fico de torta a la izquierda
axs[0].pie(
    conteo_result,
    labels=conteo_result.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['#E74C3C', '#5DADE2'],
    wedgeprops={'edgecolor': 'white'},
    textprops={'fontsize': 12}
)
axs[0].set_title('Distribuci贸n Total de Resultados', fontsize=14)

# Gr谩fico de barras a la derecha
bars = axs[1].bar(
    conteo_grupo.index,
    conteo_grupo.values,
    color=[colores[x] for x in conteo_grupo.index]
)

# Quitar ejes pero dejar t铆tulos de ejes
axs[1].set_xticks(range(len(conteo_grupo.index)))
axs[1].set_xticklabels(conteo_grupo.index, rotation=30, fontsize=10)
axs[1].set_ylabel('Cantidad de Pacientes', fontsize=12)
axs[1].set_xlabel('Grupo', fontsize=12)

# Ocultar l铆neas y marcas de ejes
axs[1].spines['top'].set_visible(False)
axs[1].spines['right'].set_visible(False)
axs[1].spines['left'].set_visible(False)
axs[1].spines['bottom'].set_visible(False)
axs[1].tick_params(axis='y', length=0)
axs[1].tick_params(axis='x', length=0)

# Etiquetas arriba de barras
for bar in bars:
    height = bar.get_height()
    axs[1].text(bar.get_x() + bar.get_width()/2, height + 5, str(int(height)),
                ha='center', va='bottom', fontsize=11)

# Crear leyenda manual con los colores y etiquetas de grupos
from matplotlib.patches import Patch
leyenda = [Patch(color=colores[k], label=k.replace('_', ' ')) for k in colores.keys()]
axs[1].legend(handles=leyenda, title='Grupos', loc='upper right')

plt.tight_layout()
plt.show()
```

##Medidas descriptivas
Se calculan estad铆sticas para variables num茅ricas y proporciones para categ贸ricas.

```python
num_cols = [
    'Age','Heart rate','Systolic blood pressure','Diastolic blood pressure','Blood sugar','CK-MB','Troponin'
            ]

desc = df[num_cols].describe().T
display(desc)

cat_cols = ['Gender', 'Result']

for col in cat_cols:
    print(f'\nDistribuci贸n de {col}')
    display(
        df[col]
          .value_counts(normalize=True)
          .rename('proporci贸n')
          .to_frame()
          .style.format('{:.1%}')
    )
```

Situaci贸n encontrada Quer铆amos saber, de un vistazo, como se comportan las variables numericas (edad, presi贸n, glucemia, enzimas) y cu谩n equilibradas est谩n las variables categ贸ricas.

Criterio elegido

Usamos describe() para obtener los n煤meros b谩sicos: cantidad de datos, promedio, dispersi贸n, m铆nimos, m谩ximos y cuartiles.

Para las categ贸ricas, calculamos la proporci贸n de cada categor铆a con value_counts(normalize=True).

Hallazgos

La edad se concentra alrededor de 58 a帽os.

La presi贸n sist贸lica media es 127 mm Hg.

El 66 % de los registros corresponden a g茅nero codificado como 1 (hombres).

El 61 % de los pacientes tienen resultado positivo en el test de ataque card铆aco.

```python
# MEDIDAS DESCRIPTIVAS BSICAS
# Columnas num茅ricas
num_cols = df.select_dtypes(include="number").columns

for col in num_cols:
    print(f" {col} (num茅rica) ")
    print("Media            :", df[col].mean())
    print("Mediana          :", df[col].median())
    print("Moda             :", df[col].mode()[0])
    print("Desv. est谩ndar   :", df[col].std())

    # An谩lisis simetr铆a
    if df[col].mean() > df[col].median():
        print(" Asimetr铆a positiva (cola a la derecha)")
    elif df[col].mean() < df[col].median():
        print(" Asimetr铆a negativa (cola a la izquierda)")
    else:
        print(" Distribuci贸n sim茅trica")
    print()

# Columnas categ贸ricas
cat_cols = df.select_dtypes(exclude="number").columns

for col in cat_cols:
    print(f" {col} (categ贸rica) ")
    print("Categor铆as 煤nicas :", df[col].nunique())
    print("Moda (m谩s repetido):", df[col].mode()[0])
    print()
```

Situaci贸n encontrada Quer铆amos resumir r谩pidamente qu茅 tan grandes, variados y balanceados son los datos num茅ricos y categ贸ricos del dataset.

Criterio elegido

Variables num茅ricas

Calculamos media, mediana, moda y desv铆o est谩ndar para ver tendencia central y dispersi贸n.

Comparamos media vs. mediana:

Si son parecidas  la distribuci贸n es sim茅trica.

Si la media es mayor  cola hacia la derecha (asimetr铆a positiva).

Si la media es menor  cola hacia la izquierda (asimetr铆a negativa).

Variables categ贸ricas

Contamos cu谩ntas categor铆as 煤nicas hay (nunique) y cu谩l se repite m谩s (mode).

Esto muestra qu茅 clase domina y si el reparto est谩 equilibrado.

Lectura simple

Una desviaci贸n est谩ndar alta indica valores muy dispersos; baja, que est谩n m谩s agrupados.

Con las proporciones de gender y result vemos, por ejemplo, que ~66 % de los registros son hombres y ~61 % tienen resultado positivo.

Saber simetr铆a y dispersi贸n nos ayuda a elegir la medida correcta (media o mediana) y a detectar posibles valores at铆picos antes de avanzar.
