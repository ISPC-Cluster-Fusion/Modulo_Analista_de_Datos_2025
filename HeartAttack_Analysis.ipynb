```python
# An√°lisis columnas num√©ricas.
# Columnas num√©ricas a analizar
columnas_numericas = ['Age', 'Heart rate', 'Systolic blood pressure', 'Diastolic blood pressure',
                      'Blood sugar', 'CK-MB', 'Troponin']

analisis_numerico = []

for columna in columnas_numericas:
    media = df[columna].mean()
    mediana = df[columna].median()
    std = df[columna].std()
    nulos = df[columna].isnull().sum()
    total = len(df[columna])

    # Valores no num√©ricos (inv√°lidos)
    registros_invalidos = df[~df[columna].apply(lambda x: isinstance(x, (int, float, np.integer, np.float64)))][columna].unique().tolist()

    analisis_numerico.append({
        "Columna": columna,
        "Valores v√°lidos": total - nulos,
        "Valores nulos": nulos,
        "Media": round(media, 2),
        "Mediana": round(mediana, 2),
        "Desviaci√≥n est√°ndar": round(std, 2),
        "Valores inv√°lidos": registros_invalidos
    })

# Convertir a DataFrame y mostrar con estilo
analisis_df = pd.DataFrame(analisis_numerico)

analisis_df.style.background_gradient(cmap='Greens').format({
    "Media": "{:.2f}",
    "Mediana": "{:.2f}",
    "Desviaci√≥n est√°ndar": "{:.2f}",
    "Valores v√°lidos": "{:.0f}",
    "Valores nulos": "{:.0f}"
})
```

```python
# An√°lisis de columnas categ√≥ricas
columnas_categoricas = ['Gender', 'Result']
analisis_categorico = []

for columna in columnas_categoricas:
    conteo = df[columna].value_counts(dropna=False).to_dict()
    valores_nulos = df[columna].isnull().sum()
    valores_invalidos = None
    if columna == 'Gender':
        valores_invalidos = df[~df[columna].isin([0, 1])][columna].unique().tolist()
    if columna == 'Result':
        valores_invalidos = df[~df[columna].isin(['positive', 'negative'])][columna].unique().tolist()
    analisis_categorico.append({
        'Columna': columna,
        'Conteo por categor√≠a': conteo,
        'Valores nulos': valores_nulos,
        'Valores fuera de lo esperado': valores_invalidos
    })

# Convertir en DataFrame
analisis_df_categorico = pd.DataFrame(analisis_categorico)

# Estilizar y mostrar Resultados
analisis_df_categorico.style.set_properties(**{'text-align': 'left'}) \
    .background_gradient(cmap='Purples') \
    .format({'Valores nulos': '{:.0f}'})
```

Criterio elegido Reemplazamos el dato vac√≠o por 58 a√±os, que es la mediana (el valor del medio al ordenar las edades). Esto evita que edades muy bajas o muy altas distorsionen el resultado y, adem√°s, no tenemos que borrar la fila.

```python
print(df.columns.tolist())
```

```python
med_Age = df['Age'].median() #mediana Age
df['Age'] = df['Age'].fillna(med_Age) #rellena valores faltantes con media

print(f"Mediana aplicada a 'Age': {med_Age}")
print("Faltantes en 'Age':", df['Age'].isna().sum())
```

Ejemplo de Dato Ausente ‚Äì Ejemplo 2

Situaci√≥n encontrada:
Se detectaron valores ausentes en la columna Age (edad). Por ejemplo, en la fila 72 no se registra la edad del paciente, lo cual es relevante para los an√°lisis de riesgo.

Criterio elegido:
Se utiliz√≥ la mediana para reemplazar el valor faltante, ya que es una medida robusta frente a valores extremos, especialmente √∫til en distribuciones no sim√©tricas como suele ser la edad.

```python
# Calcular la mediana de la edad
mediana_edad = df['Age'].median()

# Guardar los registros con edad nula antes de la imputaci√≥n
filas_con_Age_nulo = df[df['Age'].isna()].copy()  # usamos .copy() para conservar la vista previa
print("üîç Registros con edad ausente antes de la imputaci√≥n:")
display(filas_con_Age_nulo)

# Reemplazar los valores ausentes con la mediana (¬°y guardar el Resultado!)
df['Age'] = df['Age'].fillna(mediana_edad)

# Verificar cu√°ntos valores nulos quedan (deber√≠a ser 0)
print("üîé Valores nulos restantes en 'Age':", df['Age'].isna().sum())

# Mostrar los registros corregidos (usando los mismos √≠ndices anteriores)
filas_corregidas = df.loc[filas_con_Age_nulo.index]
print("\n‚úÖ Registros con edad ausente corregidos con la mediana:")
display(filas_corregidas)
```

Datos ausentes - Ejemplo 3:
Valores faltantes en resultados

```python
# Funci√≥n para inferir resultado
def inferir_resultado(fila):
    if pd.isnull(fila["Result"]):
        # Reglas de ejemplo (pueden ajustarse con base cl√≠nica m√°s precisa)
        if (
            fila["Heart rate"] > 100 or
            fila["Systolic blood pressure"] > 140 or
            fila["Blood sugar"] > 200 or
            fila["CK-MB"] > 5 or
            fila["Troponin"] > 0.01
        ):
            return "positive"
        else:
            return "negative"
    else:
        return fila["Result"]

# Aplicar la funci√≥n fila por fila
df["Result"] = df.apply(inferir_resultado, axis=1)
```

Ejemplo de Formato Incorrecto ‚Äì Ejemplo 1

Situaci√≥n encontrada:
Se detectaron valores at√≠picos o inveros√≠miles en la columna Age, espec√≠ficamente registros con edades negativas o superiores a 120 a√±os, lo cual se considera un error de formato o ingreso de datos.

Criterio elegido:
Los valores "?" se consideran err√≥neos y ser√°n tratados como datos faltantes (NaN) para luego imputarse con la mediana, ya que la mediana es menos sensible a valores extremos que el promedio.

```python
# Reemplazar edades mayores a 120 con una edad promedio razonable (80 a√±os)
df.loc[df['Age'] > 120, 'Age'] = 80

# Confirmar reemplazo
print("‚úÖ Registros corregidos con valor 80:")
display(df.loc[df['Age'] == 80].reset_index(drop=True))
```

Dato en formato incorrecto ‚Äì Ejemplo 2: Valores no num√©ricos en 'Heart rate'

Situaci√≥n encontrada:
Se detectaron registros en la columna "Heart rate" que conten√≠an valores no num√©ricos, los cuales impiden realizar an√°lisis cuantitativos.

Criterio elegido:
Se convirtieron los valores a tipo num√©rico, utilizando pd.to_numeric con errors='coerce', lo cual transforma los valores no v√°lidos en NaN. Posteriormente, los valores nulos se reemplazaron con la mediana de los valores v√°lidos, garantizando as√≠ la integridad del campo sin eliminar datos.

```python
# Buscar valores no num√©ricos en 'Heart rate'
print("üîç Valores con formato incorrecto en 'Heart rate':")
formato_incorrecto_hr = df[~df['Heart rate'].apply(lambda x: str(x).isdigit())]
display(formato_incorrecto_hr)

# Reemplazar con la mediana v√°lida
df['Heart rate'] = pd.to_numeric(df['Heart rate'], errors='coerce')
mediana_hr = df['Heart rate'].median()
df['Heart rate'] = df['Heart rate'].fillna(mediana_hr)

print(f"‚úÖ Valores corregidos en 'Heart rate' con la mediana: {mediana_hr}")
```

Formato incorrecto - Ejemplo 3:

Situaci√≥n encontrada:
Se detectaron valores abreviados e incompletos en la columna Result, tales como "neg", "pos" y valores ausentes (vac√≠os). Esta situaci√≥n puede afectar el an√°lisis, ya que impide agrupar correctamente los casos en categor√≠as consistentes.

Criterio elegido:
Se decidi√≥ reemplazar las abreviaciones "neg" por "negative" y "pos" por "positive" para estandarizar los datos. Posteriormente, se abordar√°n los valores ausentes mediante inferencia basada en otras variables cl√≠nicas del dataset (frecuencia card√≠aca, presi√≥n arterial, az√∫car en sangre, etc.).

```python
# Reemplazar valores abreviados
df["Result"] = df["Result"].replace({
    "neg": "negative",
    "pos": "positive"
})
```

Dato err√≥neo ‚Äì Ejemplo 1

Dato err√≥neo: valores negativos en "Diastolic blood pressure"
Situaci√≥n encontrada:
Se detectaron valores negativos en la presi√≥n diast√≥lica, lo cual no es fisiol√≥gicamente posible en condiciones humanas normales.

Criterio elegido:
Se reemplazaron estos valores con la mediana de los valores v√°lidos (‚â• 0), ya que representa una estimaci√≥n robusta del centro de la distribuci√≥n, evitando el sesgo de outliers o errores extremos.

```python
# Presi√≥n diast√≥lica no puede ser negativa
print("üîç Valores negativos en 'Diastolic blood pressure':")
errores_diastolica = df[df['Diastolic blood pressure'] < 0]
display(errores_diastolica)


# Reemplazar por la mediana v√°lida
mediana_diastolica = df[df['Diastolic blood pressure'] >= 0]['Diastolic blood pressure'].median()
# Convertir la columna a float para evitar conflictos con la mediana decimal
df['Diastolic blood pressure'] = df['Diastolic blood pressure'].astype(float)


# Reemplazar valores negativos por la mediana
df.loc[df['Diastolic blood pressure'] < 0, 'Diastolic blood pressure'] = mediana_diastolica
print(f"‚úÖ Reemplazados con la mediana v√°lida: {mediana_diastolica}")
```

Dato err√≥neo ‚Äì Ejemplo 2
Situaci√≥n encontrada:
Se identific√≥ un valor err√≥neo en la columna "Blood sugar" donde aparecen registros por encima de 250. En condiciones cl√≠nicas t√≠picas, un valor de glucosa en sangre por encima de 250 mg/dL puede ser considerado una hiperglucemia grave y en muchos casos extremo o improbable en contextos normales.

Criterio elegido:
Se reemplazan estos valores extremos por la mediana de los valores v√°lidos (‚â§250 mg/dL), ya que es una medida robusta ante outliers.

```python
# Buscar valores extremos de glucosa en sangre (>250)
datos_erroneos_glucosa = df[df['Blood sugar'] > 250]

print("üîç Valores err√≥neos detectados en 'Blood sugar' (mayores a 250 mg/dL):")
display(datos_erroneos_glucosa[['Blood sugar']].head(10))  # mostrar los primeros 10 para no saturar

# Reemplazar con la mediana de valores v√°lidos (‚â§250)
mediana_glucosa_valida = df[df['Blood sugar'] <= 250]['Blood sugar'].median()
df.loc[df['Blood sugar'] > 250, 'Blood sugar'] = mediana_glucosa_valida

# Mostrar confirmaci√≥n
print(f"\n‚úÖ Valores mayores a 250 reemplazados con la mediana v√°lida: {mediana_glucosa_valida}")
```

Medidas descriptivas
Se calculan estad√≠sticas para variables num√©ricas y proporciones para categ√≥ricas.

```python
num_cols = [
    'Age','Heart rate','Systolic blood pressure','Diastolic blood pressure','Blood sugar','CK-MB','Troponin'
            ]

desc = df[num_cols].describe().T
display(desc)

cat_cols = ['Gender', 'Result']

for col in cat_cols:
    print(f'\nDistribuci√≥n de {col}')
    display(
        df[col]
          .value_counts(normalize=True)
          .rename('proporci√≥n')
          .to_frame()
          .style.format('{:.1%}')
    )
```

Situaci√≥n encontrada Quer√≠amos saber, de un vistazo, como se comportan las variables numericas (edad, presi√≥n, glucemia, enzimas) y cu√°n equilibradas est√°n las variables categ√≥ricas.

Criterio elegido

Hallazgos

La edad se concentra alrededor de 58 a√±os.

La presi√≥n sist√≥lica media es 127 mm Hg.

El 66 % de los registros corresponden a g√©nero codificado como ‚Äú1‚Äù (hombres).

El 61 % de los pacientes tienen resultado positivo en el test de ataque card√≠aco.

```python
# MEDIDAS DESCRIPTIVAS B√ÅSICAS
# Columnas num√©ricas
num_cols = df.select_dtypes(include="number").columns

for col in num_cols:
    print(f"‚îÄ‚îÄ‚îÄ {col} (num√©rica) ‚îÄ‚îÄ‚îÄ")
    print("Media            :", df[col].mean())
    print("Mediana          :", df[col].median())
    print("Moda             :", df[col].mode()[0])
    print("Desv. est√°ndar   :", df[col].std())

    # An√°lisis simetr√≠a
    if df[col].mean() > df[col].median():
        print("‚Üí Asimetr√≠a positiva (cola a la derecha)")
    elif df[col].mean() < df[col].median():
        print("‚Üí Asimetr√≠a negativa (cola a la izquierda)")
    else:
        print("‚Üí Distribuci√≥n sim√©trica")
    print()

# Columnas categ√≥ricas
cat_cols = df.select_dtypes(exclude="number").columns

for col in cat_cols:
    print(f"‚îÄ‚îÄ‚îÄ {col} (categ√≥rica) ‚îÄ‚îÄ‚îÄ")
    print("Categor√≠as √∫nicas :", df[col].nunique())
    print("Moda (m√°s repetido):", df[col].mode()[0])
    print()
```

Situaci√≥n encontrada Quer√≠amos resumir r√°pidamente qu√© tan grandes, variados y balanceados son los datos num√©ricos y categ√≥ricos del dataset.

Criterio elegido

Variables num√©ricas

Calculamos media, mediana, moda y desv√≠o est√°ndar para ver tendencia central y dispersi√≥n.

Comparamos media vs. mediana:

Si son parecidas ‚Üí la distribuci√≥n es sim√©trica.

Si la media es mayor ‚Üí cola hacia la derecha (asimetr√≠a positiva).

Si la media es menor ‚Üí cola hacia la izquierda (asimetr√≠a negativa).

Variables categ√≥ricas

Contamos cu√°ntas categor√≠as √∫nicas hay (nunique) y cu√°l se repite m√°s (mode).

Esto muestra qu√© clase domina y si el reparto est√° equilibrado.

Lectura simple

Una desviaci√≥n est√°ndar alta indica valores muy dispersos; baja, que est√°n m√°s agrupados.

Con las proporciones de gender y result vemos, por ejemplo, que ~66 % de los registros son hombres y ~61 % tienen resultado positivo.

Saber simetr√≠a y dispersi√≥n nos ayuda a elegir la medida correcta (media o mediana) y a detectar posibles valores at√≠picos antes de avanzar.

### Distribuci√≥n de diagn√≠stico por g√©nero.

```python
# Distribuci√≥n de diagn√≠stico por g√©nero.

# Eliminar filas donde Gender es NaN
df = df.dropna(subset=['Gender'])

# Verificar que ya no hay nulos
print(df['Gender'].isnull().sum())


# Colores definidos antes
colores = {
    'positive_Mujer': '#F1948A',    # Rojo claro
    'positive_Hombre': '#C0392B',   # Rojo oscuro
    'negative_Mujer': '#3498DB',    # Azul claro
    'negative_Hombre': '#21618C'    # Azul oscuro
}

# Conteo total por resultado para torta
conteo_result = df['Result'].value_counts()

# Conteo por grupo para barras
# Crear columna 'Grupo' combinando Result y Gender
df["Grupo"] = df.apply(lambda fila: f"{fila['Result']}_{'Mujer' if fila['Gender'] == 0 else 'Hombre'}", axis=1)
# Conteo por grupo para barras
conteo_grupo = df['Grupo'].value_counts().reindex(colores.keys())

fig, axs = plt.subplots(1, 2, figsize=(14,6))

# Gr√°fico de torta a la izquierda
axs[0].pie(
    conteo_result,
    labels=conteo_result.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['#E74C3C', '#5DADE2'],  # verde y morado claros para positivo y negativo
    wedgeprops={'edgecolor': 'white'},
    textprops={'fontsize': 12}
)
axs[0].set_title('Distribuci√≥n Total de Resultados', fontsize=14)

# Gr√°fico de barras a la derecha
bars = axs[1].bar(
    conteo_grupo.index,
    conteo_grupo.values,
    color=[colores[x] for x in conteo_grupo.index]
)

# Quitar ejes pero dejar t√≠tulos de ejes
axs[1].set_xticks(range(len(conteo_grupo.index)))
axs[1].set_xticklabels(conteo_grupo.index, rotation=30, fontsize=10)
axs[1].set_ylabel('Cantidad de Pacientes', fontsize=12)
axs[1].set_xlabel('Grupo', fontsize=12)

# Ocultar l√≠neas y marcas de ejes
axs[1].spines['top'].set_visible(False)
axs[1].spines['right'].set_visible(False)
axs[1].spines['left'].set_visible(False)
axs[1].spines['bottom'].set_visible(False)
axs[1].tick_params(axis='y', length=0)
axs[1].tick_params(axis='x', length=0)

# Etiquetas arriba de barras
for bar in bars:
    height = bar.get_height()
    axs[1].text(bar.get_x() + bar.get_width()/2, height + 5, str(int(height)),
                ha='center', va='bottom', fontsize=11)

# Crear leyenda manual con los colores y etiquetas de grupos
from matplotlib.patches import Patch
leyenda = [Patch(color=colores[k], label=k.replace('_', ' ')) for k in colores.keys()]
axs[1].legend(handles=leyenda, title='Grupos', loc='upper right')

plt.tight_layout()
plt.show()
```

### G√©nero: Distribuci√≥n de Pacientes y Promedio de Presi√≥n Sist√≥lica

```python
# G√©nero: Distribuci√≥n de Pacientes y Promedio de Presi√≥n Sist√≥lica.

# Etiquetas de g√©nero
df["Gender"] = df["Gender"].replace({1: "Masculino", 0: "Femenino"})

# Datos para gr√°fico 1: Conteo por g√©nero
gender_counts = df["Gender"].value_counts().reindex(["Masculino", "Femenino"])

# Datos para gr√°fico 2: Promedio de presi√≥n sist√≥lica
promedios = df.groupby("Gender")["Systolic blood pressure"].mean().round(1).reindex(["Masculino", "Femenino"])

# Colores (m√°s contrastantes pero consistentes)
colores_genero = {
    "Masculino": "#6BAED6",  # azul claro
    "Femenino": "#BC80BD"    # naranja suave
}

# Crear figura
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Ocultar las l√≠neas de los ejes (bordes)
for spine in axs[0].spines.values():
    spine.set_visible(False)

for spine in axs[1].spines.values():
    spine.set_visible(False)

# Ocultar las marcas (ticks) de los ejes
axs[0].tick_params(left=False, bottom=False)
axs[1].tick_params(left=False, bottom=False)

# Saco escalas de los ejes.
axs[0].set_yticklabels([])
axs[1].set_yticklabels([])

axs[0].tick_params(axis='y', length=0)
axs[1].tick_params(axis='y', length=0)



# Subplot 1: Distribuci√≥n por g√©nero
axs[0].bar(gender_counts.index, gender_counts.values,
           color=[colores_genero[gen] for gen in gender_counts.index])
axs[0].set_title("Distribuci√≥n por G√©nero")
axs[0].set_ylabel("Cantidad de Pacientes")

# Etiquetas arriba
for i, val in enumerate(gender_counts.values):
    axs[0].text(i, val + 5, str(int(val)), ha='center', va='bottom', fontsize=10)

# Subplot 2: Promedio presi√≥n sist√≥lica por g√©nero
bars = axs[1].bar(promedios.index, promedios.values,
                  color=[colores_genero[gen] for gen in promedios.index])
axs[1].set_title("Presi√≥n Sist√≥lica Promedio por G√©nero")
axs[1].set_ylabel("Presi√≥n Sist√≥lica (mmHg)")
axs[1].set_ylim(110, 130)
axs[1].grid(axis='y', linestyle="--", alpha=0.5)

# Etiquetas arriba de las barras
for i, val in enumerate(promedios.values):
    axs[1].text(i, val + 0.5, f"{val} mmHg", ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

Este gr√°fico doble permite observar la proporci√≥n de pacientes seg√∫n g√©nero, y luego comparar los valores promedio de presi√≥n sist√≥lica entre hombres y mujeres. Se mantienen los colores usados previamente (azul para hombres, violeta para mujeres) para asegurar consistencia visual y claridad.

### Tabla Comparativa de Promedios Cl√≠nicos por Diagn√≥stico
